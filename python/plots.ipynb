```python
alg_list = []
thread_list = []
ur_list = []
size_list = []
throughput_list = []

with open("output.txt", "r") as file:
    for line in file:
        parts = line.split()
        if len(parts) == 11:
            alg_list.append(parts[1])
            thread_list.append(int(parts[3]))
            ur_list.append(int(parts[5]))
            size_list.append(int(parts[7]))
            throughput_list.append(float(parts[10]))
```

```python
data = {
    'ALG': alg_list,
    'THREAD': thread_list,
    'UR': ur_list,
    'SIZE': size_list,
    'Throughput (ops/s)': throughput_list
}

df = pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.ticker as mticker

def set_scientific_yaxis(ax):
    formatter = mticker.ScalarFormatter(useMathText=True)
    formatter.set_powerlimits((0, 0))
    ax.yaxis.set_major_formatter(formatter)

sns.set(style="whitegrid")

df1 = df[:54]
df2 = df[54:108]
df3 = df[108:]

for alg in df1['ALG'].unique():
    data = df1[df1['ALG'] == alg]
    plt.figure(figsize=(10, 6))
    for size in data['SIZE'].unique():
        sub_data = data[data['SIZE'] == size]
        sns.lineplot(data=sub_data, x='THREAD', y='Throughput (ops/s)', label=size)
    plt.title(f'{alg} with Fixed Update Ratio and Varying List Size')
    plt.xlabel('Threads')
    plt.ylabel('Throughput (ops/s)')
    plt.legend(title='List Size')
    set_scientific_yaxis(plt.gca())
    plt.savefig(f'charts/{alg}_ur10.png')
    plt.close()

for alg in df2['ALG'].unique():
    data = df2[df2['ALG'] == alg]
    plt.figure(figsize=(10, 6))
    for ur in data['UR'].unique():
        sub_data = data[data['UR'] == ur]
        sns.lineplot(data=sub_data, x='THREAD', y='Throughput (ops/s)', label=f'{ur}%')
    plt.title(f'{alg} with Fixed List Size and Varying Update Ratio')
    plt.xlabel('Threads')
    plt.ylabel('Throughput (ops/s)')
    plt.legend(title='Update Ratio')
    set_scientific_yaxis(plt.gca())
    plt.savefig(f'charts/{alg}_ls100.png')
    plt.close()

plt.figure(figsize=(10, 6))
for alg in df3['ALG'].unique():
    data = df3[df3['ALG'] == alg]
    sns.lineplot(data=data, x='THREAD', y='Throughput (ops/s)', label=alg)
plt.title('All Algorithms with Fixed Update Ratio and List Size')
plt.xlabel('Threads')
plt.ylabel('Throughput (ops/s)')
plt.legend(title='Algorithm')
set_scientific_yaxis(plt.gca())
plt.savefig('charts/3algs_ur10_ls1000.png')
plt.close()
```
